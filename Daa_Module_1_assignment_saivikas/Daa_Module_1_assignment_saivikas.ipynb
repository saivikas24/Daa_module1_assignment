```python
import random
import time
import sys
def bubble_sort(arr):
    n = len(arr)
    sort = 0
    for i in range(n):
        for j in range(0, n-i-1):
            sort += 1  # Increment operation count for each comparison
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                sort = sort+ 1  
    return arr, sort
    
def insertion_sort(arr):
    sort = 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            sort =sort +1 
            arr[j+1] = arr[j]
            sort = sort +1
            j -= 1
        arr[j+1] = key
        sort = sort +1
    return arr, sort

def selection_sort(arr):
    sort = 0
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            sort = sort +1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            sort = sort +1
    return arr, sort
```

```python
def analyze_sort(sort_func, arr):
    arr_copy = arr.copy()
    start_time = time.time()
    sorted_arr, operations = sort_func(arr_copy)
    end_time = time.time()
    time_taken = end_time - start_time
    memory_used = sys.getsizeof(arr_copy) + sys.getsizeof(sorted_arr)  # Estimate memory usage by summing sizes
    return time_taken, operations, memory_used
```

```python
size = 100
array = [random.randint(1, 1000) for _ in range(size)]

algorithms = {
    "Bubble": bubble_sort,
    "Insertion": insertion_sort,
    "Selection": selection_sort
}
for name, func in algorithms.items():
    time_taken, sort, memory = analyze_sort(func, array)
    print(f"{name}: Time={time_taken:.6f}s, Operations={sort}, Memory={memory} bytes")
```
